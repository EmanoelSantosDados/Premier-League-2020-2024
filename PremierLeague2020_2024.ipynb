## **Excluindo colunas**
Essas colunas não tem relevância para esse trabalho.

```python
df.drop(columns=['Unnamed: 0', 'time', 'comp', 'xg', 'attendance', 'xga', 'captain', 'match report',
              'notes', 'sot', 'dist', 'fk', 'pk', 'pkatt'], axis=1, inplace=True)
```

```python
df.dtypes
```

## **Tratando as colunas**

```python
df["date"] = pd.to_datetime(df["date"], format='%Y-%m-%d', errors='coerce')  #Convertendo a data

#convertendo essas colunas em categoria para economizar memória e operações mais rápidas
df["venue"]=df["venue"].astype("category")
df["day"]=df["day"].astype("category")
df["result"]=df["result"].astype("category")
df["opponent"]=df["opponent"].astype("category")
df["team"]=df["team"].astype("category")
df["round"]=df["round"].astype("category")
df["referee"]=df["referee"].astype("category")
df["formation"]=df["formation"].astype("category")
df["season"]=df["season"].astype("category")

#convertendo dados em bits mais baixos
df["gf"]=df["gf"].astype(np.int8)
df["sh"]=df["sh"].astype(np.int8)

df.head()
```

```python
df.info()
```

```python
premier = df.copy()
```

```python
premier.columns
```

```python
#Excluindo a palavra Matchweek
premier['round'] = premier['round'].str.replace('Matchweek', '')
premier
```

## **Traduzindo para português**

```python
#Traduzindo colunas português
premier_col = {'date': 'data',
               'day': 'dia',
               'round': 'rodada',
               'venue': 'local',
               'referee': 'arbitro',
               'formation': 'formação_mand',
               'team': 'time',
               'opponent': 'visitante',
               'result': 'resultado',
               'gf': 'gols_mand',
               'ga': 'gols_visit',
               'sh': 'chutes_mand',
               'poss': 'posse_mand',
               'poss.1': 'posse_visit',
               'season': 'temporada',

}

#Traduzindo linhas português
premier_lin = {'Mon': 'Segunda',
               'Tue': 'Terça',
               'Wed': 'Quarta',
               'Thu': 'Quinta',
               'Fri': 'Sexta',
               'Sat': 'Sábado',
               'Sun': 'Domingo',
               'W': 'Vitória',
               'D': 'Empate',
               'L': 'Derrota',
               'Home': 'Mandante',
               'Away': 'Visitante',
}
```

```python
premier.rename(columns=premier_col, inplace=True)
premier.replace(premier_lin, inplace=True)

premier.copy()
```

## **Criando colunas**

```python
#Criando a coluna placar
premier['placar'] = premier['gols_mand'].astype(str) + '-' + premier['gols_visit'].astype(str)

#Criando a coluna posse_vist
premier['posse_visit'] = 100 - premier['posse_mand']

premier
```

```python
#Ordenando as colunas
premier = premier[['data', 'dia', 'rodada', 'local', 'arbitro', 'formação_mand', 'time',
                   'visitante', 'resultado', 'placar', 'gols_mand', 'gols_visit', 'posse_mand',
                   'posse_visit', 'chutes_mand', 'temporada']
]
premier.head()
```

## **Gráficos**

```python
#Total de gols por rodada
total_gols = premier['gols_mand'] + premier['gols_visit']

plt.figure(figsize=(12, 6))
sns.barplot(x='rodada', y=total_gols, data=premier, palette='viridis')
plt.title('Goals Distribuição por rodada')
plt.xlabel('Rodada')
plt.ylabel('Gols')
plt.xticks(rotation=45)

plt.show()
```

```python
#Total de gols por temporada
plt.figure(figsize=(12, 6))
sns.barplot(x='temporada', y=total_gols, data=premier, palette='viridis')
plt.title('Goals Distribuição por temporada')
plt.xlabel('Temporada')
plt.ylabel('Gols')
plt.xticks(rotation=45)

plt.show()
```

```python
#Resultados por temporadas
sns.countplot(data=premier, x="resultado",hue="temporada")
```

```python
#Resultados por dia
sns.countplot(data=premier, x="resultado", hue="dia")
```

```python
#Número de vitórias por dia
vitoria = premier.groupby("dia")["data"].count().reset_index(name='count')
sns.barplot(data=vitoria, x='dia', y='count', palette='viridis')
plt.xlabel('Dia da semana')
plt.ylabel('número de vitórias')
plt.title('Número de vitórias por dia da semana')

plt.show()
```

```python
#Times com mais vitórias
vitoria = premier[premier["resultado"]=="Vitória"].groupby("time")["resultado"].count().sort_values(ascending=False)
vitoria.plot(kind="bar")
```

```python
#Times com mais vitórias por temporada
premier2 = premier.copy()
premier2['Vitória'] = premier2['resultado'].apply(lambda x: 1 if x == 'Vitória' else 0)

temp_premier = premier2.pivot_table(index='time', columns='temporada', values='Vitória', aggfunc='sum')

plt.figure(figsize=(15, 15))
sns.heatmap(temp_premier, annot=True, linewidth=0.5, cmap='summer')
```

```python

temp_premier = premier2.pivot_table(index='time', columns='temporada', values='gols_mand', aggfunc='sum')

plt.figure(figsize=(15, 15))
sns.heatmap(temp_premier, annot=True, linewidth=0.5, cmap='summer',fmt="d")
```

```python
#Total partidas - juiz
arbitro = premier2.groupby("arbitro")["data"].count().sort_values(ascending=False)

arbitro.plot(kind="bar")
```
